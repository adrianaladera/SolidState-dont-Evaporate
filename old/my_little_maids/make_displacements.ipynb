```python
import os
import shutil
import subprocess as sp
from pymatgen.core import Structure
from pymatgen.io.vasp import Poscar
```

```python
path = "/Users/adrianaladera/Desktop/MIT/research/mochas/ALL_STRUCTURES/mochas_opendac/original_files/"

for file in os.listdir(path):
    mocha_dir = file.split(".")[0]
    if file.endswith(".vasp"):
        if not os.path.exists(f"{path}{mocha_dir}"):
            os.mkdir(f"{path}{mocha_dir}")
            shutil.copyfile(f"{path}{file}", f"{path}{mocha_dir}/POSCAR")
    elif file.endswith(".cif"):
        structure = Structure.from_file(f"{path}{file}")
        if not os.path.exists(f"{path}{mocha_dir}"):
            os.mkdir(f"{path}{mocha_dir}")
        Poscar(structure).write_file(f"{path}{mocha_dir}/POSCAR")
```

```python
for file in os.listdir(path):
    if os.path.isdir(f"{path}{file}"):
        os.chdir(f"{path}{file}")
        sp.run("phonopy -d --dim=\"1 1 1\"", shell=True)
        print(file)
        os.chdir(path)
```

```python
for mocha in os.listdir(path):
    if os.path.isdir(f"{path}{mocha}") and mocha != "original_files":
        for file in os.listdir(f"{path}{mocha}"):
            if "POSCAR" in file and file != "SPOSCAR" and file != "POSCAR":
                os.rename(f"{path}{mocha}/{file}", f"{path}{mocha}/{mocha}{file[-4:]}")
                print(f"{path}{mocha}/{mocha}{file[-4:]}")
            elif "POSCAR" in file and file != "SPOSCAR" and file == "POSCAR":
                os.rename(f"{path}{mocha}/{file}", f"{path}{mocha}/{mocha}")
                print(f"{path}{mocha}/{mocha}")
            elif "SPOSCAR" == file or "disp.yaml" in file:
                os.rename(f"{path}{mocha}/{file}", f"{path}{mocha}/{mocha}_{file}")
                print(f"{path}{mocha}/{mocha}_{file}")
```
